# Evaluation

```python
def evaluation(predicted_rul, true_rul):
    #a1 = 13
    #a2 = 10
    negative = 0
    for i in range(0, len(predicted_rul)):
        if (predicted_rul[i] < 0):
            predicted_rul[i] = 0
            negative = negative + 1
    d = predicted_rul - true_rul    
    pred = pd.DataFrame(predicted_rul)
    true = pd.DataFrame(true_rul)
    result = pd.concat([true, pred], axis=1)
    result.columns = (['true', 'pred'])
    result['diff'] = (result['true'] - result['pred'])
    score = 0
    for i in range(0, len(d)):
        if (d[i]<0):
#             x = np.exp(-(d[i]/13))-1
            x = -(d[i]/13)
            
        else:
#             x = np.exp(d[i]/10)-1
            x = d[i]/10
        score = score + x
    score_per_test = score/len(predicted_rul)
    print('Negative predictions: ', negative)
    print('Score: ', score)
    print('AVG Score per test: ', score_per_test)    
    return(result)
```

```python
def mean_prediction_score(RUL_real, RUL_pred):
    d = RUL_pred - RUL_real
    
    return (np.sum(np.exp(d[d >= 0] / 13) - 1) +
            np.sum(np.exp(-1 * d[d < 0] / 10) - 1)) / len(RUL_real)
```

```python
def Onprediction(predicted_rul,true_rul):
    gamma1=1
    gamma2=2
    RUL_prozent=[]
    d = predicted_rul - true_rul    
    pred = pd.DataFrame(predicted_rul)
    true = pd.DataFrame(true_rul)
    result = pd.concat([true, pred], axis=1)
    result.columns = (['true', 'pred'])
    result['diff'] = (result['true'] - result['pred'])
    score = 0
    for i in range(0,len(true_rul)):
        t_diff = (len(true_rul)-i)/len(true_rul)
        if true_rul[i]==0:
            true_rul[i]=1
            
        d = 1-np.abs((predicted_rul[i]-true_rul[i])/true_rul[i])
        x = gamma1*t_diff+gamma2*d
        score = score + x
        RUL_prozent.append(t_diff*100)
#         x = gamma1*t_diff
#         x = gamma2*d
    result['RUL %']=RUL_prozent

        
    print('Score: ', score)       
    return result
```
